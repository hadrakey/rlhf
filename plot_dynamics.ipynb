```python
import pandas as pd
```

```python
df=pd.read_csv("dpo_acc_gpt1.csv")
df1=pd.read_csv("dpo_acc_gpt2.csv")
df2=pd.read_csv("dpo_acc_gpt3.csv")
df3=pd.read_csv("dpo_acc_gpt4.csv")
```

```python
DF=pd.concat([df,df1,df2,df3])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(DF["size"], DF.acc, linestyle = 'dotted')
plt.xlabel("size of preference data (training)")
plt.ylabel("rewards/accuracies")
plt.title("DPO on Anthropic Helpful-Harmless")
plt.show()
```
